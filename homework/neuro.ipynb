```python
#this file was written using edf2ascii converter http://www.teuniz.net/edf2ascii/
#on the first edf test file http://www.teuniz.net/edf_bdf_testfiles/index.html
data = loadtxt('/home/jeforero/Downloads/ma0844az_1-1+_data.txt', delimiter=',')
```

```python
time = data[:,0]
signal_1 = data[:,1]
```

```python
n_times = size(time)
n_sample=400 #new number of times that I want to have
n_jump = n_times/n_sample #size that I have to jump in each signal
n_signals = 24 #total number of signals I want to keep
```

```python
all_signals = ones((n_sample, n_signals))
```

```python
for i in range(n_signals):
    for j in range(n_sample):
        all_signals[j,i] = data[j*n_jump,i+1]
```

```python
for i in range(20,n_signals):
    plot(all_signals[:,i])
```

```python
fileout = open('sampled+ma0844az_1-1+_data.txt','w')
for i in range(n_sample):
    fileout.write('%s\n' %(' '.join([str(j) for j in all_signals[i,:]])))
  #  fileout.write(all_signals[i,:])
fileout.close()
```
