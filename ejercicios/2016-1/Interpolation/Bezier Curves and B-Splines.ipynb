## Bézier Curves

The Quadratic Bézier curve is a function of $t$, where t ranges from 0 to 1.

$\mathbf{B}(t) = (1 - t)^{2}\mathbf{P}_0 + 2(1 - t)t\mathbf{P}_1 + t^{2}\mathbf{P}_2 \mbox{ , } t \in [0,1].$

Where $\mathbf{P}_n$ are the control points. So if we define the function in a way we can use it, and then define some values for the control points:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def B(t):
    return (1-t)**2*p0 + 2*t*(1-t)*p1 + t**2*p2
```

```python
p0 = np.array([0, 0])
p1 = np.array([1.5, 1])
p2 = np.array([2, 0])
```

```python
for t in np.linspace(0, 1, 20):
    zz= B(t)
zz
```

```python
xy = np.array([B(t) for t in np.linspace(0, 1, 20)]).T
xy
```

```python
plt.plot(xy[0], xy[1])
```

The Cubic Bézier curve has the form of:
$\mathbf{R}(t)=(1-t)^3\mathbf{P}_0+3(1-t)^2t\mathbf{P}_1+3(1-t)t^2\mathbf{P}_2+t^3\mathbf{P}_3 \mbox{ , } t \in [0,1].$
Again, we can define the function and plot it for a given set of points:

```python
def R(t,p0,p1,p2,p3):
    return (1-t)**3 * p0 + 3 * t * (1-t)**2 * p1 + 3 * t**2 * (1-t) * p2 + t**3 * p3
```

```python
p3 = np.array([1, -2])
xy = np.array([R(t,p0,p1,p2,p3) for t in np.linspace(0, 1, 30)]).T
plt.plot(xy[0], xy[1])
```

# B-Splines

We can concatenate different Bèzier curves in a way that ensures continuity and smoothness.

```python
def Q(t, p0, p1, p2, p3):
    return (((1-t)**3)/(6)) * p0 + ((3*t**3 -6*t**2 + 4)/(6)) * p1 + ((-3*t**3 + 3*t**2 + 3*t + 1)/(6)) * p2 + ((t**3)/(6)) * p3
```

```python
p4 = np.array([-1, -1])
p5 = np.array([1, 3])
```

If we now use our 6 points and plot 4 at a time and keep shifting in one new control point then the spline closes itself.

```python
xy = np.array([Q(t, p0, p1, p2, p3) for t in np.linspace(0, 1, 30)]).T
plt.plot(xy[0], xy[1])
xy = np.array([Q(t, p1, p2, p3, p4) for t in np.linspace(0, 1, 30)]).T
plt.plot(xy[0], xy[1])
xy = np.array([Q(t, p2, p3, p4, p5) for t in np.linspace(0, 1, 30)]).T
plt.plot(xy[0], xy[1])
xy = np.array([Q(t, p3, p4, p5, p0) for t in np.linspace(0, 1, 30)]).T
plt.plot(xy[0], xy[1])
xy = np.array([Q(t, p4, p5, p0, p1) for t in np.linspace(0, 1, 30)]).T
plt.plot(xy[0], xy[1])
xy = np.array([Q(t, p5, p0, p1, p2) for t in np.linspace(0, 1, 30)]).T
plt.plot(xy[0], xy[1])
```

B-splines are a great, portable method for interpolating not-too noisy data. For this we use splrep and splev from scipy.interpolate. Those functions are built on top of much older FORTRAN routines. See the function documentation for more info.

```python
from scipy import interpolate 
```

```python
x = np.linspace(0.0,10.0,21)
y = np.sin(x**1.5 + 2.0)
x_true=np.linspace(0.0,10.0,1000)
y_true = np.sin(x_true**1.5 + 2.0)
plt.scatter(x,y)
```

The B-Spline interpolation algorithm yields a tuple containing the knots vector, the B-Spline coefficients (related to the control points) and the degree of each Bézier curve. See Suggested Reading above for further clarification.

```python
tck  = interpolate.splrep(x, y)
tck
```

The tuple above can be used in other languages (MATLAB, Mathematica, R) to construct the interpolating curve. You may have to fiddle with the tck-tuple format in order to port it successfully, though.

```python
y_spline = interpolate.splev(x_true,tck)
```

```python
plt.scatter(x,y)
plt.plot(x_true, y_spline)
plt.plot(x_true, y_true)
plt.legend(['true', 'spline'])
```

The function splev can also yield the derivative of the interpolating curve. Notice what happens when we calculate the second derivative.

```python
y_prime = interpolate.splev(x_true,tck,der=1)
y_dprime = interpolate.splev(x_true,tck,der=2)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.plot(x_true, y_spline)
plt.plot(x_true, y_prime)
plt.plot(x_true, y_dprime)
plt.legend(['spline','1st der','2nd der'])
```
