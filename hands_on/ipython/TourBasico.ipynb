#Histogramas

```python
a = random.random(100000)
histo, bin_edges = histogram(a, bins=100)
```

```python
plot(bin_edges[1:], histo)
```

```python
width = 0.5*(bin_edges[1]-bin_edges[0])
center = (bin_edges[1:] + bin_edges[:-1]) * 0.5
bar(center, histo, align = 'center', width=width)
```

```python
n, bins, patches = hist(a, 50, normed=1, facecolor='green', alpha=0.80)
```

```python
a = random.standard_normal(100000)
histo, bin_edges = histogram(a, bins=100)
```

```python
plot(bin_edges[1:], histo)
```

```python
n, bins, patches = hist(a, 50, normed=1, facecolor='green', alpha=0.80)
```
