```python
encephalograms = loadtxt("../../ComputationalMethodsData/homework/hw_4/sampled+ma0844az_1-1+_data.txt")
plot(encephalograms[:,20])
```

```python
shape(encephalograms)
```

```python
n_electrodes = shape(encephalograms)[1]
n_dimensions = shape(encephalograms)[0]
for i in range(4):
    plot(encephalograms[:,i])
```

```python
mean_values = zeros(n_dimensions)
for i in range(n_dimensions):
    mean_values[i] = mean(encephalograms[i,:])
```

```python
plot(mean_values)
plot(encephalograms[:,0])
```

```python
cov_matrix = zeros((n_dimensions,n_dimensions))
for i in range(n_dimensions):
    for j in range(n_dimensions):
        cov_matrix[i,j] =  sum((encephalograms[i,:] - mean_values[i]) * (encephalograms[j,:] - mean_values[j]))/(n_electrodes-1.0)
```

```python
imshow(log10(abs(cov_matrix))+1E-30)
```

```python
eigenvalues, eigenvectors = la.eig(cov_matrix)
```

```python
print (eigenvalues[0]), eigenvalues[-1]
```
