#Bayesian parameter estimation

The central aim in many branches in science is to estimate $N$ parameters describing a physical model: $\Theta=(\Theta_1, \Theta_2,\ldots,\Theta_N)$ from a set of $M$ measurements, $D=(D_1, \ldots, D_M)$ with a variance $\sigma=(\sigma_1, \ldots, \sigma_M)$. This is what we did in the linear algebra section of this course.

Here we will present another mathematical rigorous way to make this estimation. The big difference is that this time we will try to describe the *belief* we have on our model using a probabilistic approach.

Using Bayes theorem we introduce the term likelihood $L\equiv L(D|\Theta)$ which can be read as the probability of having the data points $D$ given a certain value of the parameters $\Theta$. Assiming that each data point $D_i$ is distributed as a gaussian with variance $\sigma_i$, and that the data points are mutually independent, one can show that the likelihood can be written as


$$
L \propto \prod_{i=0}^{M} \exp \left[ -\frac{1}{2} \left(\frac{D_i-D}{\sigma_i}\right)^2\right]
=\exp \left[ -\frac{1}{2} \sum_{i=0}^{M}\left(\frac{D_i-D}{\sigma_i}\right)^2\right] = \exp{\left(-\frac{1}{2}\chi^2\right)}
$$

where

$$
\chi^2 = \sum_{i=0}^{M}\left(\frac{D_i-D}{\sigma_i}\right)^2
$$

A good way to estimate the best parameters in the model is to sample the likelihood using the Metropolis-Hastings algorithm. 

Below is a detailed example of this procedure in the case of a set points modeled by a line.

```python
%pylab inline
#load the list of points
obs_data = loadtxt("obs_data.dat")
x_obs = obs_data[:,0]
y_obs = obs_data[:,1]
```

```python
scatter(x_obs,y_obs)
```

```python
def likelihood(y_obs, y_model):
    chi_squared = sum((y_obs-y_model)**2)
    return exp(-chi_squared)
```

```python
def my_model(x_obs, m, b):
    return x_obs*m + b
```

```python
m_walk = empty((0)) #this is an empty list to keep all the steps
b_walk = empty((0))

m_walk = append(m_walk, random.random())
b_walk = append(b_walk, random.random())
print m_walk
print b_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    m_prime = np.random.normal(m_walk[i], 0.1) 
    b_prime = np.random.normal(b_walk[i], 0.1)

    y_init = my_model(x_obs, m_walk[i], b_walk[i])
    y_prime = my_model(x_obs, m_prime, b_prime)

    alpha = likelihood(y_obs, y_prime)/likelihood(y_obs, y_init)
    if(alpha>=1.0):
        m_walk  = append(m_walk,m_prime)
        b_walk  = append(b_walk,b_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            m_walk = append(m_walk,m_prime)
            b_walk = append(b_walk,b_prime)
        else:
            m_walk = append(m_walk,m_walk[i])
            b_walk = append(b_walk,b_walk[i])
```

```python
scatter(m_walk, b_walk)
```

```python
count, bins, ignored =plt.hist(m_walk, 20, normed=True)
```

```python
count, bins, ignored =plt.hist(b_walk, 20, normed=True)
```

```python
average_m = average(m_walk)
average_b = average(b_walk)
```

```python
best_y = my_model(x_obs, average_m, average_b)
scatter(x_obs,y_obs)
plot(x_obs, best_y)
```
